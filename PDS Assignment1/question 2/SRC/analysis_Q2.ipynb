Data visualization between genders (male and female) and race.
This graph has two colors, showing clear differences between males and females.
The x-axis shows race, and the y-axis shows the number of people there.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data =  pd.read_csv("clean_data.csv")


df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='race/ethnicity', hue='gender')
plt.title('Distribution of Race/Ethnicity by Gender')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Count')
plt.show()
```

Data visualization between gender and parental level of education.
The different colors show the different levels of education of the parents.
The x-axis shows the parent's level of education, and the y-axis shows the number of people there.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame called df
# First, let's create a DataFrame with counts of each gender and parental level of education
gender_parent_education_counts = df.groupby(['gender', 'parental level of education']).size().reset_index(name='count')

# Now, let's create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='parental level of education', y='count', hue='gender', data=gender_parent_education_counts)
plt.title('Count of Gender by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-labels for better readability
plt.tight_layout()
plt.show()
```

Data visualization of gender and lunch distribution.
The different colors show differences between males and females.
In the x-axis it represents gender, and in the y-axis, it represents the number of them.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame called df
# First, let's create a count plot
plt.figure(figsize=(8, 6))
sns.countplot(x='lunch', hue='gender', data=df)
plt.title('Gender Distribution by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

Data visualization between gender and test preparation course.
The X-axis represents the test preparation course, and the Y-axis represents the number of them.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame called df
# First, let's create a count plot
plt.figure(figsize=(8, 6))
sns.countplot(x='test preparation course', hue='gender', data=df)
plt.title('Gender Distribution by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

Data visualization of gender and math score.
In the X-axis it represents gender, and in the Y-axis it represents math score.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame called df
# First, let's create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Math Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame called df
# First, let's create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='reading score', data=df)
plt.title('Reading Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.tight_layout()
plt.show()
```

A bar plot makes it simpler to analyze the link between gender and test preparation course since it makes it easy to compare the number of students in each group. A box plot makes it easier to understand how reading scores vary by gender by highlighting any disparities in central tendency and variability between male and female.
