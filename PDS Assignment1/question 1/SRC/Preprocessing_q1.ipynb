```python
import pandas as pd
raw_dataset=pd.read_csv('/content/raw_data - Sheet1.csv')
```

```python
#preprocessing the data in preparation for analysis
#updating the column names with the relevant units of measurement
raw_dataset.columns=['Height(in)', 'Weight(lb)', 'Age', 'Grip_strength(kg)','Frailty']
raw_dataset.columns
# Converting categorical data to numerical class values
raw_dataset['Frailty'].replace(['Y', 'N'], [0, 1], inplace=True)
```

```python
clean_data=raw_dataset.dropna()
clean_data
```

```python
import pandas as pd
from google.colab import files
df = pd.DataFrame(clean_data)
df.to_csv('clean_data.csv', index=False)
files.download('clean_data.csv')
```
